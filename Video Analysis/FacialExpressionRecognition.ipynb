### Facial Expression Recognition from Image

```python
from fer import FER
import cv2
```

```python
img = cv2.imread("test_images_fer/test_img_1.jpg")
emotion_detector = FER()
if(not emotion_detector.detect_emotions(img)):
    emotion_detector = FER(mtcnn=True)

# emotion_detector.detect_emotions(img)
emotion, confidence = emotion_detector.top_emotion(img)
confidence*=100
print("Facial Expression : "+emotion.title())
print("Confidence Level : {:0.2f} %".format(confidence))
    
```

### Facial Expression Recognition from Video

```python
from fer import FER
from fer import Video
import os
import sys
```

```python
video_file = "test_videos_fer/test_video_2.mp4"
video = Video(video_file)

emotion_detector = FER(mtcnn=True)
raw_data = video.analyze(emotion_detector, display=False)

df = video.to_pandas(raw_data)
df = video.get_first_face(df)
df = video.get_emotions(df)
df
```

```python
print("Facial Expression : "+ df.mean(axis=0).idxmax().title())
print("Confidence Level : {:0.2f} %".format(df.mean(axis=0).max()*100))
```
